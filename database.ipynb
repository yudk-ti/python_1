```python
# 전자제품 데이터 입출력용 SQLite 클래스 예제
import sqlite3
import random

class ElectronicsDB:
    def __init__(self, db_path=':memory:'):
        self.conn = sqlite3.connect(db_path)
        self.cur = self.conn.cursor()
        self.create_table()

    def create_table(self):
        self.cur.execute("""
            CREATE TABLE IF NOT EXISTS Electronics (
                id INTEGER PRIMARY KEY,
                name TEXT,
                price INTEGER
            )
        """)
        self.conn.commit()

    def insert(self, id, name, price):
        self.cur.execute(
            "INSERT INTO Electronics (id, name, price) VALUES (?, ?, ?)",
            (id, name, price)
        )
        self.conn.commit()

    def update(self, id, name=None, price=None):
        if name is not None and price is not None:
            self.cur.execute(
                "UPDATE Electronics SET name=?, price=? WHERE id=?",
                (name, price, id)
            )
        elif name is not None:
            self.cur.execute(
                "UPDATE Electronics SET name=? WHERE id=?",
                (name, id)
            )
        elif price is not None:
            self.cur.execute(
                "UPDATE Electronics SET price=? WHERE id=?",
                (price, id)
            )
        self.conn.commit()

    def delete(self, id):
        self.cur.execute("DELETE FROM Electronics WHERE id=?", (id,))
        self.conn.commit()

    def select(self, id=None):
        if id is None:
            self.cur.execute("SELECT * FROM Electronics")
            return self.cur.fetchall()
        else:
            self.cur.execute("SELECT * FROM Electronics WHERE id=?", (id,))
            return self.cur.fetchone()

# 샘플 데이터 100개 생성 및 삽입
db = ElectronicsDB('electronics_sample.db')
product_names = [f"제품{i}" for i in range(1, 101)]
for i, name in enumerate(product_names, 1):
    price = random.randint(10000, 1000000)
    db.insert(i, name, price)

# CRUD 예시
print('전체 데이터 개수:', len(db.select()))
print('1번 제품:', db.select(1))
db.update(1, name='수정된제품', price=55555)
print('수정 후 1번 제품:', db.select(1))
db.delete(1)
print('삭제 후 1번 제품:', db.select(1))
```
